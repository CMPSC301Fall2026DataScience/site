## Part 2: Your First Data Analysis

Let's create a simple dataset and analyze it:

```python
# Create sample data
np.random.seed(42)

data = {
    'Student': [f'Student {i}' for i in range(1, 11)],
    'Math': np.random.randint(70, 100, 10),
    'Science': np.random.randint(65, 95, 10),
    'English': np.random.randint(75, 100, 10)
}

df = pd.DataFrame(data)

print("📊 Student Grades Dataset:")
print(df)
print("\n📈 Summary Statistics:")
print(df.describe())
```

## Part 3: Data Manipulation

Let's perform some common data operations:

```python
# Calculate average grade for each student
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)

# Add a grade category
df['Grade'] = pd.cut(df['Average'], 
                      bins=[0, 70, 80, 90, 100],
                      labels=['C', 'B', 'A', 'A+'])

print("📊 Enhanced Dataset:")
print(df)

print("\n🏆 Top 3 Students:")
print(df.nlargest(3, 'Average')[['Student', 'Average', 'Grade']])
```

## Part 4: Visualization

Let's create a bar chart:

```python
# Create visualization
plt.figure(figsize=(10, 6))

x = range(len(df))
width = 0.25

plt.bar([i - width for i in x], df['Math'], width, label='Math', color='#FF6B6B')
plt.bar(x, df['Science'], width, label='Science', color='#40C9D9')
plt.bar([i + width for i in x], df['English'], width, label='English', color='#6BCB77')

plt.xlabel('Students')
plt.ylabel('Grades')
plt.title('Student Performance Across Subjects', fontsize=14, fontweight='bold')
plt.xticks(x, df['Student'], rotation=45)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("✅ Visualization created successfully!")
```
